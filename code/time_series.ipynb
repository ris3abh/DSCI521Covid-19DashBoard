```python
def cumulative_sum(df):
    result_df = df.copy()
    a = result_df.iloc[2:, 1:]
    for i in range(1, len(a.columns)):
        a.iloc[:, i] = a.iloc[:, i] + a.iloc[:, i-1]
    return a
```

### Check for stationarity:

```python
import pandas as pd

# Load the dataset into a pandas DataFrame

# Sum the values for each column to obtain the total number of cases for each date

total_df = result_df[1:].sum(axis=0)
```

```python
## converting series to dataframe
total_df = pd.DataFrame(total_df, columns=['total'])
```

```python
total_df.head()
```

```python
#In time series analysis, it is important to check for stationarity of the data, which means the statistical properties of the data do not change over time. There are several methods to check for stationarity, including the Augmented Dickey-Fuller (ADF) test and the KPSS test.

from statsmodels.tsa.stattools import adfuller

def adf_test(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
adf_test(total_df)
```

```python
## detrnding the data
total_df['detrend'] = total_df['total'] - total_df['total'].rolling(7).mean()

## plotting the detrended data
total_df['detrend'].plot(figsize=(12, 6))
```

From the trend we can predict that covid-19 cases were increasing exponentially, but after the lockdown the cases started to decrease.
January 2020 was the first lockdown. The other major lockdown was in march of 2021 where we can see a sudden drop in the cases. The cases started to increase again after the lockdown was lifted. The cases are increasing exponentially again in the end of 2021. With this rise mexican government made it compulsary to use masks in public places and weak masks are banned. The cases are decreasing again after the lockdown. On 2022-04-04 Mexico was marked 'green' and then a month later again to 'yellow'. In the November of 2022 masks were again made mandatory when indoors. This trend can also be seen in our analysis after detrending the data.

### The data is not stationary, so we need to make it stationary. We can do this by taking the difference of the data. The difference of the data is the data minus the previous data. This will remove the trend from the data. We can also use the log of the data to remove the trend. We can also use the difference of the log of the data to remove the trend and seasonality.

```python
adf_test(total_df['detrend'].dropna())
```

## we can say over here, that the data is stationary as the p value for the data is less than 0.05

```python
## ARIMA model (Autoregressive Integrated Moving Average) or an exponential smoothing model.

from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")

train = total_df.iloc[:int(0.8*(len(total_df))), :]
test = total_df.iloc[int(0.8*(len(total_df))):, :]
print(train.shape, test.shape)

## fitting the model
model = ARIMA(train['total'], order=(1, 1, 1))
model_fit = model.fit()
print(model_fit.summary())
```

model appears to fit the data well, as evidenced by the high Log Likelihood value and the significant coefficients for the autoregressive (AR) and moving average (MA) terms. The p-values for the AR and MA coefficients are both less than 0.05, indicating that they are statistically significant at a 95% confidence level. The Ljung-Box test statistic for the first lag is also low, indicating that there is no significant autocorrelation in the residuals.

The value of the Heteroskedasticity (H) test statistic is high, which suggests that there may be some heteroscedasticity in the data, meaning that the variance of the residuals may not be constant over time. However, this can be common in time series data and may not necessarily affect the accuracy of the model.

Overall, the ARIMA(1, 1, 1) model seems like a reasonable choice for modeling the data, but it's important to keep in mind that the appropriateness of the model depends on the specific characteristics of the data and the goals of the analysis. It's also a good practice to validate the model's assumptions and evaluate its performance on out-of-sample data before making any conclusions.

```python
## predicting the values
start_index = test.index[0]
end_index = test.index[-1]
predictions = model_fit.predict(start=1, end= len(test), typ='levels').rename('ARIMA Predictions')
```

```python
## plotting the predictions
title = 'Total Cases'
ylabel='Cases'
xlabel=''
ax = test['total'].plot(legend=True, figsize=(12,6), title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x', tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel)

from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse

error = rmse(test['total'], predictions)
print(f'RMSE Error: {error:11.10}')
```
